```python
data = {
    'geo': ['London', 'Michigan', 'Lisbon', 'Copenhagen'],
    'collections': [34, 29, 40, 11]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['collections'], labels=df['geo'], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Collections by Region')

plt.savefig('collections_by_region.png', format='png', bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```
